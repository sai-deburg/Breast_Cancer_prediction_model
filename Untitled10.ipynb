```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
# data=data.drop('who',axis=1)
# data=data.drop('deck',axis=1)
data=data.drop('parch',axis=1)
```

```python
data
```

```python
data['embark_town'].value_counts()
```

```python
data['embark_town']=data['embark_town'].map({'Southampton':0,'Cherbourg':1,'Queenstown':2})
```

```python
data=data.drop('deck',axis=1)
```

```python
data['embarked']=data['embarked'].map({'S':0,'C':1,'Q':2})
```

```python
data['sex']=data['sex'].map({'male':0,'female':1})
```

```python
data
```

```python
data=data.drop('adult_male',axis=1)
data['alive']=data['alive'].map({'no':0,'yes':1})
data['class']=data['class'].map({'first':1,'second':2,'third':0})
```

```python
data=data.drop('who',axis=1)
```

```python
data['alone']=data['alone'].map({'False':0,'True':1})
```

```python
data=data.drop('class',axis=1)
data=data.drop('alone',axis=1)
```

```python
data
```

```python
X=data[['pclass','sex','age','sibsp','fare','embarked','embark_town','alive']]
```

```python
X = X.dropna(subset=['age'])
X.isna().sum()

```

```python
y=data['survived']
```

```python

X = X.dropna(subset=['embarked'])
X = X.dropna(subset=['embark_town'])
X.isna().sum()
```

```python
y
```

```python
X
data1=pd.concat([X, y], axis=1)
data1=data1.dropna()
data1
X=data1.drop('survived',axis=1)
y=data1['survived']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l2'],'C':[1,2,3,4,5,6,10,20,30],'max_iter':[100,200,300]}
classifier_Regretor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_Regretor.fit(X_train,y_train)
```

```python
print(classifier_Regretor.best_params_)
```

```python
print(classifier_Regretor.best_score_)
```

```python
y_pred=classifier_Regretor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy=accuracy_score(y_pred,y_test)
```

```python
print(accuracy)
```

```python
print(classification_report(y_pred,y_test))
```
